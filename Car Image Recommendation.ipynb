```python
def load_data(path= r"/content/drive/MyDrive/AmberAlert2/dataset"):
    data = {}
    images = []
    names = []

    columns = ['Make', 'Model', 'Year']

    for i in columns:
        data[i] = []
        
    for name in os.listdir(path):
        names.append(name)

        name = name.split(".")[0]

        for col, val in zip(columns, name.split("_")):
            data[col].append(val)
        
            
    df = pd.DataFrame(data)
    df["name"] = names
    
    return df
```

```python
df = load_data()
```

```python
df.to_csv("all_cars.csv", index=False)
```
